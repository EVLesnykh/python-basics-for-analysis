```python
def my_range(start, end):
    current = start
    while current < end:
        yield current
        current += 1
```

```python
ranger = my_range(0, 2)
ranger
```

```python
next(ranger)
```

```python
gen_a = my_range(2, 5)
for i in gen_a:
    print(i)
```
